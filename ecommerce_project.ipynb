```python
def slab_c(value):
    if value<=0.5:
        return 0.5
    elif 0.5 <value <=1:
        return 1
    elif 1<value<=1.5:
        return 1.5
    elif 1.5 <value <= 2:
        return 2
    elif 2< value <= 2.5:
        return 2.5
    elif 2.5< value<= 3:
        return 3
    elif 3< value <= 3.5:
        return 3.5
    elif 3.5< value<= 4:
        return 4
final['weight_slab_c'] = final['Charged Weight'].map(slab_c)
final.head()
```
